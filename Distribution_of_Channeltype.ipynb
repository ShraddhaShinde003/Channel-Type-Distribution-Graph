# Task : Create a function to calculate the distribution of channeltype from the top 1000 records

```python
# Import Python Libraries
import numpy as np
import pandas as pd
import mysql.connector as sql
from sqlalchemy import create_engine
import pymysql
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Dataset
df = pd.read_csv('youtube_dataset.csv')

# Drop rows having Null values in ChannelType column
df=df.dropna(subset=['channeltype'])
```

```python
# Create a filtered dataframe with first 1000 rows of youtube_dataset
df_filtered = df.iloc[:1000,:]

# Convert Categorical data from ChannelType column into numerical form
df_filtered['channeltype'] = df_filtered['channeltype'].astype('category')
df_filtered['channeltype'] = df_filtered['channeltype'].cat.codes
```

```python
# Create a Function 
def distribution(df):
    df_filtered.channeltype.plot.density(color='green')
```

```python
# Call a Function
distribution(df_filtered)
```
